```
def sort(a,aux,lo,hi):
  
  if lo >= hi:return 

  mean = (lo+hi)*0.5
  mid = int(mean)

  sort(a=a,aux=aux,lo=lo,hi = mid)
  sort(a=a,aux=aux,lo = mid+1,hi = hi)
  merge(a=a,aux=aux,lo=lo,hi=hi,mid=mid)
  
  plot_bar(a)
  
```

```
def merge(a,aux,lo,hi,mid):
  

  for k in range(lo,hi+1):
    aux[k]=a[k]
    
  i = lo
  j = mid+1  
  for k in range(lo,hi+1):
    if(i>mid):
      a[k] = aux[j]
      j+=1
    elif(j>hi):
      a[k] = aux[i]
      i+=1
    elif(aux[i]<aux[j]):
      a[k] = aux[i]
      i+=1
    else :
      a[k]=aux[j]
      j+=1
    
```

```
def plot_bar(arr):
  x = np.arange(len(arr))
  y = arr
  plt.title("Merge Sort")
  plt.xlabel("Index")
  plt.ylabel("Value")
  plt.bar(x,y)
  plt.show()
```
