```
import matplotlib.pyplot as plt
import numpy as np
import random
```

```
def isSorted(a):
  for i in range(1,len(a)):
    if a[i-1]>a[i]: return False
  return True  
```

```
duplicate = [30]*6
list = [*range(40,30,-1)] + duplicate + [*range(30,0,-1)]
desc = np.array(list)
desc
```

```
def exch(a,i,j):
  tmp = a[i]
  a[i] = a[j]
  a[j] = tmp
```

```
def shuffle(arr):
  for i in range(len(arr)-1,0,-1):
    r = random.randint(0,i)
    exch(arr,i,r)
```

```
def exch(a,i,j):
  tmp = a[i]
  a[i] = a[j]
  a[j] = tmp
```

```
def sorted(a):
  plot_bar(a,"Unsorted Array")
  shuffle(a)
  plot_bar(a,"Knuth Shuffled")
  hi = len(a)-1
  lo = 0
  sort(a=a,lo=lo,hi=hi)
  assert isSorted(a),"Unsorted."
  
```

```
def sort(a,lo,hi):
  
  if hi<=lo:return 
  if(a[hi]<a[lo]):exch(a,lo,hi)
  pivot = partition(a,lo,hi)
  
  sort(a=a,lo=lo,hi=pivot-1)
  sort(a=a,lo=pivot+1,hi=hi)
  plot_bar(a,"QuickSort")
  if isSorted(a):return
```

```
def partition(a,lo,hi):

  partition_element_index = lo
  j = hi
  i = lo+1
  while True:

    while(a[i]<=a[partition_element_index]):
      i+=1
      if i == hi:break
    while(a[partition_element_index]<a[j]):
      j-=1
      if j == lo:break;
    if(i>=j): break;
    exch(a,i,j) 
  exch(a,partition_element_index,j)
  return j  
```

```
def plot_bar(arr,title):
  x = np.arange(len(arr))
  y = arr
  plt.title(title)
  plt.xlabel("Index")
  plt.ylabel("Value")
  plt.bar(x,y)
  plt.show()
```
