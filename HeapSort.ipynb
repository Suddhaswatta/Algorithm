```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
def plot_bar(a,title):
  x = np.arange(len(a))
  y = a
  plt.bar(x,y)
  plt.xlabel("Index")
  plt.ylabel("Values")
  plt.title(title)
  plt.show()
```

```
def sink(a,k,N):
  
  while(2*k<N-1):
    j = 2*k+1
    if (j<N-1) and (a[j+1]>a[j]) : j=j+1
    if a[k]>a[j] : break
    print(f'parent: {a[k]}, parent index:{k} child:{a[j]} child index :{j}')
    exch(a,k,j)
    k = j
 

```

```
def exch(a,i,j):
  tmp = a[i]
  a[i] = a[j]
  a[j] = tmp
```

```
def heapify(a):
  N = len(a)
  k = int(N/2)
  for i in range(k,-1,-1):
    sink(a,i,N)  
```

```
a = np.array([30,29,28,54,45,78,96,100,120,130,90,70,71,17,19,31,65,31])
heapify(a)
a
```

```
def sort(a):

  plot_bar(a,"Unsorted Array")
  heapify(a)
  N = len(a)
  while(N>0):
    
    exch(a,0,N-1)
    N-=1
    sink(a,0,N)
    
    
    plot_bar(a,"Heap Sort")
   
```
