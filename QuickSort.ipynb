```
def exch(a,i,j):
  tmp = a[i]
  a[i] = a[j]
  a[j] = tmp
```

```
def shuffle(arr):
  for i in range(len(arr)-1,0,-1):
    r = random.randint(0,i)
    exch(arr,i,r)
```

```
def exch(a,i,j):
  tmp = a[i]
  a[i] = a[j]
  a[j] = tmp
```

```
def sorted(a):
  plot_bar(a,"Array in Descending Order")
  shuffle(a)
  plot_bar(a,"Knuth Shuffled")
  hi = len(a)-1
  lo = 0
  sort(a=a,lo=lo,hi=hi)
  
```

```
def sort(a,lo,hi):
  
  if lo >= hi:return 
  plot_bar(a,"QuickSort")
  pivot = partition(a,lo,hi)
  sort(a=a,lo=lo,hi=pivot-1)
  sort(a=a,lo=pivot+1,hi=hi)

  
  
```

```
def partition(a,lo,hi):

  j = hi
  i = lo
  while True:

    while(a[i]<a[lo]):
      i+=1
    while(a[j]>a[lo]):
      j-=1
    if(i>=j): break;
    exch(a,i,j) 
  exch(a,lo,j)
  return j  
```

```
def plot_bar(arr,title):
  x = np.arange(len(arr))
  y = arr
  plt.title(title)
  plt.xlabel("Index")
  plt.ylabel("Value")
  plt.bar(x,y)
  plt.show()
```
