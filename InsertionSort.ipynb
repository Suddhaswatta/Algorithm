```
def plot_bar(arr,title):
    x = np.arange(len(arr))
    y = np.array(arr)
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.title(title)
    plt.bar(x,y)
    plt.show()
```

Check if array is sorted helper

Swap Helper

```
def exch(a,i,j):
  tmp = a[i]
  a[i]=a[j]
  a[j]=tmp
```

Insertion Short Algorithm

```
def sort(arr):
  plot_bar(arr,"Unsorted Array")
 
  N = len(arr)
  for i in range(1,N):
    for j in range(i,0,-1):
      if(arr[j]<arr[j-1]):
        exch(arr,j,j-1)
    plot_bar(arr,"Insertion Sort")
```
